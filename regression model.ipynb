```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Input feature
X = np.array([10, 20, 30, 40, 50, 60]).reshape(-1, 1)

# Target values (Perfect Linear Relationship: y = 2x + 5)
y = np.array([25, 45, 65, 85, 105, 125])

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Performance Metrics
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
print("Mean Squared Error:", mse)
print("R² Score:", r2)

# Predict for a new value
new_value = np.array([[70]])
prediction = model.predict(new_value)
print("Prediction for X = 70:", prediction[0])

# Plot
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')
plt.title("Linear Regression - Perfect Fit")
plt.xlabel("Advertising Expenditure")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```
